```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('1694953404-Mall_Customers.csv')
df.head()
```

```python
df.columns
```

```python
plt.scatter(df['Age'],df['Spending Score (1-100)'])
plt.xlabel('age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
plt.scatter(df["Age"],df["Annual Income (k$)"])
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
```

```python
relevant_cols = ["Age", "Annual Income (k$)", 
                 "Spending Score (1-100)"]

customer_df = df[relevant_cols]
```

```python
customer_df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(customer_df)
```

```python
scalar_data=scalar.transform(customer_df)
```

```python
scalar_data
```

```python
def find_best_clusters(df, maximum_K):
    clusters_centers = []
    k_values = []
    for k in range(2, maximum_K):
        kmeans_model = KMeans(n_clusters = k)
        kmeans_model.fit(df)

        clusters_centers.append(kmeans_model.inertia_)
        k_values.append(k)

    return clusters_centers, k_values
```

```python
clusters_centers, k_values = find_best_clusters(scalar_data, 12)
```

```python
def generate_elbow_plot(clusters_centers, k_values):
    
    figure = plt.subplots(figsize = (12, 6))
    plt.plot(k_values, clusters_centers, 'o-', color = 'orange')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("Cluster Inertia")
    plt.title("Elbow Plot of KMeans")
    plt.show()
```

```python
generate_elbow_plot(clusters_centers, k_values)
```

```python
kmeans_model = KMeans(n_clusters = 5)
```

```python
kmeans_model.fit(scalar_data)
```

```python
KMeans(n_clusters=5)
```

```python
df["clusters"] = kmeans_model.labels_
```
